```
boston=sklearn.datasets.load_boston()
x=boston.data
y=boston.target
columns=boston.feature_names
```

```
type(x)
columns
```

```
df=pd.DataFrame(x)
df.columns=columns
df
```

```
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
IQR
```

```
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
lower_bound,type(upper_bound)
```

```
df<lower_bound
```

```
cleaned_df=df[(df.CRIM.values>lower_bound['CRIM']) & (df.CRIM.values<upper_bound['CRIM'])]
cleaned_df
```

```
cd=df[~(df<lower_bound).any(axis=1)]
cd=cd[~(cd>upper_bound).any(axis=1)]
cd
```

```
((df<lower_bound) | (df>upper_bound)).any(axis='columns')
```

```
Final_clean=df[((df<lower_bound) | (df>upper_bound)).any(axis='columns')]
Final_clean.head()
```
